## 6. Guardar resultados y exportar el notebook

```python
import zipfile
from google.colab import files

os.makedirs('/content/work/outputs', exist_ok=True)

# Copiar salidas a Drive
if os.path.exists('/content/work/outputs'):
    !cp -r /content/work/outputs/* "$DRIVE_OUT" 2>/dev/null || true

zip_path = '/content/work/outputs_suicidios.zip'
with zipfile.ZipFile(zip_path, 'w', zipfile.ZIP_DEFLATED) as zf:
    if os.path.exists('/content/work/outputs'):
        for root, _, filenames in os.walk('/content/work/outputs'):
            for name in filenames:
                full = os.path.join(root, name)
                rel = os.path.relpath(full, '/content/work')
                zf.write(full, rel)

print('ZIP generado:', zip_path)
files.download(zip_path)
```
